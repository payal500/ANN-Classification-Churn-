```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
```

```python
df=pd.read_csv('Churn_Modelling.csv')
df.head()
```

```python
df=df.drop(['RowNumber','CustomerId','Surname'],axis=1)
df
```

```python
#label encoder to convert gender 
gender=LabelEncoder()
df['Gender']=gender.fit_transform(df['Gender'])
df
```

```python
from sklearn.preprocessing import OneHotEncoder
geography=OneHotEncoder()
geo_encoder=geography.fit_transform(df[['Geography']])
geo_encoder
geo_code_df=pd.DataFrame(geo_encoder.toarray(),columns=geography.get_feature_names_out(['Geography']))
geo_code_df
data=pd.concat([df.drop('Geography',axis=1),geo_code_df],axis=1)
data
```

```python
with open('gender.pkl','wb') as file:
    pickle.dump(gender,file)
with open('geography.pkl','wb') as file:
    pickle.dump(geography,file)
```

```python
#train test split
x=data.drop(['EstimatedSalary'],axis=1)
y=data['EstimatedSalary']
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
with open('scaler.pkl','wb')as file:
    pickle.dump(scaler,file)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import TensorBoard,EarlyStopping
import datetime
```

```python
# build model
Model=Sequential([
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)),
    Dense(32,activation='relu'),
    Dense(1)
])
```

```python
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01) # we cn define optimizers like this or also directoly optimizer='Adam'
# we can efine lossed like this or directly like loss='binar_crossentropy'
##complie the model

Model.compile(optimizer=opt,loss='mean_absolute_error',metrics=['accuracy'])
```

```python
Model.summary()
```

```python
#setup the tensorboard 
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
log_dir='regressionlogs/fit/' + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
#set up early stopping 
early_stopping_calling=EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
#train the model
history=Model.fit(
    x_train,y_train,validation_data=(x_test,y_test),epochs=100,
    callbacks=[tensorflow_callback,early_stopping_calling]
)
```

```python
Model.save('regression_model.h5')

```

```python
#define a function to create model and try diff parameters(keras classifier)

def create_model(neurons=32,layers=1):
    Model=Sequential()
    Model.add(Dense(neurons,activation='relu',input_shape=(x_train.shape[1],)))

    for _ in range(layers-1):
        Model.add(Dense(neurons,activation='relu'))

    Model.add(Dense(1))
    Model.compile(optimizer='adam',loss='binary_classentropy',metrics=['accuracy'])
    return Model
```

```python
#create a keras classiffier 
Model=KerasClassifier(layer=1,neurons=32,build_fn=create_model,verbose=1)
```

```python
#define grid search parameters
param_grid={
    'neurons':[16,32,64,128],
    'layers':[1,2],
    'epochs':[50,100]
}
```

```python
#perform grid search 

# grid=GridSearchCV(estimator=Model,param_grid=param_grid,n_jobs=-1,cv=3,verbose=1)
# grid_result=grid.fit(x_train, y_train)
grid = GridSearchCV(estimator=Model, param_grid=param_grid, n_jobs=-1, cv=3, verbose=1)
grid_result = grid.fit(x_train, y_train)
print("Best Accuracy: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
#print the parameters
# print("best: %f using %s" % (grid_result.best_score_,grid_result.best_params_))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam, RMSprop
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import GridSearchCV

# Define a function that builds your model
def create_model(optimizer='adam'):
    model = Sequential()
    model.add(Dense(12, input_dim=x_train.shape[1], activation='relu'))  # update input_dim if needed
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model

# ✅ Wrap the model
model = KerasClassifier(model=create_model, verbose=0)

# ✅ Define the parameter grid
param_grid = {
    'model__optimizer': ['adam', 'rmsprop'],
    'batch_size': [10, 20],
    'epochs': [10, 50]
}

# ✅ Use GridSearchCV
grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=-1, cv=3, verbose=1)
grid_result = grid.fit(x_train, y_train)

# ✅ Show results
print("Best Accuracy: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```
